```python
import os
import numpy as numpy
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import load_img, img_to_array
from dotenv import load_dotenv

load_dotenv()
# Create a directory to save the augmented images
output_dir="C:\\Users\\nleo\\Downloads\\Output\\Interacoustics_Lyra"
os.makedirs(output_dir, exist_ok=True)

# Load your image
image_path = "C:\\Users\\nleo\\Downloads\\lyra-58cb9ab7.webp"
image=load_img(image_path)
image_array = img_to_array(image)

image_array = image_array.reshape((1,) + image_array.shape)

datagen=ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
)
# Uncomment datagen if you want to be able to reproduce the exact images (reproducibility)
datagen.random_seed=42

i=0
# Change the save_prefix to change your filename
for batch in datagen.flow(image_array, batch_size=1, save_to_dir=output_dir, 
                          save_prefix='Interacoustics_Lyra', save_format='png'):
    if i < 5:
        plt.figure(i)
        imgplot = plt.imshow(batch[0].astype('uint8'))
    i += 1
    # Set the break point here
    if i >=100:
        break

plt.show()
```
